```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.metrics import confusion_matrix
```

```python
def fit_perceptron(X_train, y_train):
    # Add  column of 1
    X_train_new = np.hstack((np.ones((X_train.shape[0], 1)), X_train))

    # Initialize w to 0
    w = np.zeros((X_train.shape[1]+1,1))

    epoch = 0

    #use the errorPer as firstcheck to make sure if all weight is zero is also ok
    while epoch < 5000 and errorPer(X_train_new, y_train, w) != 0:

        for i in range(len(X_train_new)):

            if pred(X_train_new[i], w) != y_train[i]:
               #isue of matrix size need to reshape
                w_change = X_train_new[i] * y_train[i]
                w += np.reshape(w_change, (X_train.shape[1]+1, 1))

        epoch+=1

    return w
```

```python
def errorPer(X_train,y_train,w):
  N = np.shape(X_train)[0]

  num = 0.0

  for i in range(N):
    if y_train[i] != pred(X_train[i],w):
      num+=1

  return num/N
```

```python
def confMatrix(X_train,y_train,w):
  matrix = np.zeros((2,2))
  N = np.shape(X_train)[0]

  x_1 = np.ones((N, 1))
  X_train = np.hstack((np.ones((N, 1)), X_train))

  for i in range (N):

    pred_label = pred(X_train[i], w)

    if pred_label ==  y_train[i]:
      if pred_label == 1:
        #true postive
        matrix[1][1] += 1
      else:
        #true negative
         matrix[0][0] += 1
    else:
      if pred_label == 1:
        #true postive
        matrix[0][1] += 1
      else:
        #false negative
         matrix[1][0] += 1

  return matrix
```

```python
def pred(X_train,w):
  result = np.dot(X_train,w)
  if result > 0:
    return 1
  else:
    return -1
```

```python
def test_SciKit(X_train, X_test, Y_train, Y_test):

    pct = Perceptron(max_iter=5000)
    pct.fit(X_train, Y_train)

    pred_pct = pct.predict(X_test)

    return confusion_matrix(Y_test, pred_pct)
```

```python
def test_Part1():
    from sklearn.datasets import load_iris
    X_train, y_train = load_iris(return_X_y=True)
    X_train, X_test, y_train, y_test = train_test_split(X_train[50:],y_train[50:],test_size=0.2)

    #Set the labels to +1 and -1
    y_train[y_train == 1] = 1
    y_train[y_train != 1] = -1
    y_test[y_test == 1] = 1
    y_test[y_test != 1] = -1

    #Pocket algorithm using Numpy
    w=fit_perceptron(X_train,y_train)
    cM=confMatrix(X_test,y_test,w)

    #Pocket algorithm using scikit-learn
    sciKit=test_SciKit(X_train, X_test, y_train, y_test)

    #Print the result
    print ('--------------Test Result-------------------')
    print("Confusion Matrix is from Part 1a is: ",cM)
    print("Confusion Matrix from Part 1b is:",sciKit)


test_Part1()
```
