```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

```python
def fit_LinRegr(X_train, y_train):
   #make robust to singular matrices
   ones = np.ones((X_train.shape[0], 1))
   robust = np.hstack((ones, X_train))
   #least squares formula
   trans_X =  robust.transpose()
   pseudoInverse = np.dot(trans_X, robust)
   nextStep = np.dot(np.linalg.pinv(pseudoInverse), trans_X)
   w = np.dot(nextStep, y_train)
   return w
```

```python
def mse(X_train,y_train,w):
    # call pred function
    prediction = pred(X_train, w)
    # calculate mean
    return np.mean(((y_train - prediction) ** 2))
```

```python
def pred(X_train,w):
    #make robust to singular matrices
    ones = np.ones((X_train.shape[0], 1))
    robust = np.hstack((ones, X_train))

    return np.dot(robust, w)
```

```python
def test_SciKit(X_train, X_test, Y_train, Y_test):
    #import model
    model = linear_model.LinearRegression()
    #train model
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    return mean_squared_error(Y_test, Y_pred)
```

```python
def subtestFn():
    # This function tests if your solution is robust against singular matrix

    # X_train has two perfectly correlated features
    X_train = np.asarray([[1, 2], [2, 4], [3, 6], [4, 8]])
    y_train = np.asarray([1,2,3,4])

    try:
      w=fit_LinRegr(X_train, y_train)
      print ("weights: ", w)
      print ("NO ERROR")
    except:
      print ("ERROR")
```

```python
def testFn_Part2():
    X_train, y_train = load_diabetes(return_X_y=True)
    X_train, X_test, y_train, y_test = train_test_split(X_train,y_train,test_size=0.2)

    w=fit_LinRegr(X_train, y_train)

    #Testing Part 2a
    e=mse(X_test,y_test,w)

    #Testing Part 2b
    scikit=test_SciKit(X_train, X_test, y_train, y_test)

    print("Mean squared error from Part 2a is ", e)
    print("Mean squared error from Part 2b is ", scikit)

print ('------------------subtestFn----------------------')
subtestFn()

print ('------------------testFn_Part2-------------------')
testFn_Part2()
#the performance of my implementation is 11 decimal places identical to the existing scikit-learn function output
```
